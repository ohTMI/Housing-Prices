# Univariate Analysis

### Let's start with our Numerical Features

```python
n = len(num_feats)
cols = 4
rows = (n + cols - 1 ) // cols

fig, axes = plt.subplots(rows, cols, figsize=(cols * 5, rows * 4))

axes = axes.flatten()

for i, col in enumerate(num_feats):
    sns.histplot(data[col].dropna(), kde=True, bins=30, ax=axes[i])
    axes[i].set_ylabel('Number of Houses')
    axes[i].grid(True)

for ax in axes[n:]:
    fig.delaxes(ax)

plt.tight_layout()
plt.show()
```

# Bivariate Analysis

#### While we were able to make observations about which features are redundant in our univariate analysis, lets see if any other features could be further dropped to streamline our modeling by comparing the Top 25 features of both Numerical and Categorical to 'SalePrice'

```python
num_data = data.select_dtypes(include=['number'])
corr = num_data.corr()
top_corr = corr['SalePrice'].abs().sort_values(ascending=False)[1:26] # Top 25 feautures exlcuding 'SalePrice'
top_corr_feats = top_corr.index.tolist()

plt.figure(figsize=(16, 14))
sns.heatmap(data[top_corr_feats + ['SalePrice']].corr(), annot=True, cmap= 'coolwarm', fmt='.2f')
plt.title('Top 25 Features Correlated with SalePrice')
plt.show()
```
